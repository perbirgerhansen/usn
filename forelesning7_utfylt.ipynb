<h2 style="color:darkblue;">if-elif-else</h1>

elif minner om else, men er snevrere ved at elif har en betingelse.

Vi prøver! Vi skriver følgende kode i den tomme kodecellen nedenfor.

```python
x = 10
y = 0

if (x > y):
    print('x er større enn y. if-grenen er aktiv.')
elif (x == y):
    print('x er identisk med y. elif-grenen er aktiv.')
else:
    print('x er ikke større enn y og heller ikke lik y. else-grenen er aktiv.')

print('Kodelinje etter if-elif-else-strukturen.')
```

<h2 style="color:darkblue;">elif kan brukes til klassifisering</h1>

Eksempel:

Universitets- og høgskolerådet (UHR) har utarbeidet følgende standard poeng-karakterskala for eksamener:

* [90, 100] = A
* [80, 90> = B
* [60, 80> = C
* [50, 60> = D
* [40, 50> = E
* [0, 40> = F

Firkantparentes betyr "inkludert". Vinkelparentes betyr "ikke inkludert".

Nedenfor er et program som beregner karakter (k) fra gitt poengsum (p). (Jeg hadde opprinnelig tenkt at vi skulle kode dette sammen, men jeg så at det ble for liten tid, så jeg/vi spurte KI-verktøyet Copilot i stedet, se lenger ned i notebooken.)

```python
p = 39

if (100 >= p >= 90):
    k = "A"
elif (90 > p >= 80):
    k = "B"
elif (80 > p >= 60):
    k = "C"
elif (60 > p >= 50):
    k = "D"
elif (50 > p >= 40):
    k = "E"
else:
    k = "F"
    
print(k)
```

Jeg spurte Copilot om følgende:

-----------------------------

Universitets- og høgskolerådet (UHR) har utarbeidet følgende standard poeng-karakterskala for eksamener:

[90, 100] = A
[80, 90> = B
[60, 80> = C
[50, 60> = D
[40, 50> = E
[0, 40> = F

Kan du lage et Python-program som beregner karakter fra poengsum?


```python
# Kode servert av Copilot:

def beregn_karakter(poengsum):
    if 90 <= poengsum <= 100:
        return "A"
    elif 80 <= poengsum < 90:
        return "B"
    elif 60 <= poengsum < 80:
        return "C"
    elif 50 <= poengsum < 60:
        return "D"
    elif 40 <= poengsum < 50:
        return "E"
    elif 0 <= poengsum < 40:
        return "F"
    else:
        return "Ugyldig poengsum"

# Eksempel på bruk
poengsum = float(input("Skriv inn poengsum: "))
karakter = beregn_karakter(poengsum)
print(f"Karakter: {karakter}")
```

Vi ser at koden inneholder en if-elif-else-struktur, her programmert i funksjonen beregn_karakter.
