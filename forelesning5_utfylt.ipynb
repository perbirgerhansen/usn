<h1 style="color:darkred;">Hvordan f en funksjon til returnere flere verdier enn n verdi?</h1>

Vi studerer og kjrer flgende program, som bruker en tuppel til  returnere flere verdier.

```python
def fun_verdi_og_stigning(x, a, b):
    y = a*x + b
    dy = a
    return (y, dy)

p0 = 2.5
p1 = 3.0
inn = 10.0

(verdi, stigning) = fun_verdi_og_stigning(inn, p0, p1)
print('Funksjonsverdi = ', verdi)
print('Stigningstall = ', stigning)
```

Programmet ovenfor brukte en tuppel til returnere flere enn n verdi. Alternativt kan vi bruke liste eller array. La oss prve begge ved  modifisere ovennevnte program.

```python
# Vi legger det modifiserte programet inn i denne kodecellen.
def fun_verdi_og_stigning(x, a, b):
    y = a*x + b
    dy = a
    return [y, dy]

p0 = 2.5
p1 = 3.0
inn = 10.0

L = fun_verdi_og_stigning(inn, p0, p1)  # L blir en liste med to elementer siden funksjonen
                                        # returnerer en liste med to elementer.
verdi = L[0]
stigning = L[1]
print('Funksjonsverdi = ', verdi)
print('Stigningstall = ', stigning)
```

<h1 style="color:darkred;">Funksjonskall med standardargumenter</h1>

Antall faktiske innargumenter ("kall-argumenter") kan vre frre enn antall formelle innargumenter (som du angir i funksjonsdefinisjonen). De manglende faktiske argumentene m du i s fall ha definert som standardargumenter med forhndsverdier. Kryptisk? La oss studere et eksempel!

```python
def fun_lin(x, a, b=3.0):
    y = a*x + b
    return y

x_value = 10.0
p0 = 2.5

out = fun_lin(x_value, p0)
print('Result = ', out)
```

<h1 style="color:darkred;">Funksjonskall med keyword-argument</h1>

Med keyword-argumenter spiller rekkeflgen av argumentene **ingen** rolle nr vi kaller funksjonen.

```python
def fun_lin(x, a, b):
    y = a*x + b
    return y

inn = 10.0
p0 = 2.5
p1 = 3.0

# Vanlig funksjonskall:
ut1 = fun_lin(inn, p0, p1)
print('ut1 = ', ut1)

# Funksjonskall med keyword-argumenter:
ut2 = fun_lin(a=p0, b=p1, x=inn)
print('ut2 = ', ut2)
```

<h1 style="color:darkred;">Dokumentasjonstekst (docstring)</h1>

Du kan legge inn det vi norsk kan kalle dokumentasjonstekst (pythonsk: docstring) i funksjoner som du lager. Dokumentasjonstekst er tekst der du kan beskrive funksjonens oppbygning, dens inn- og ut-argumenter og deres datatyper, og hva funksjonen gjr. Dokumentasjonsteksten kan vises blant annet med koden help(funksjonsnavnet).

```python
def fun_lin(x, a, b):
    """Beregning av verdien av liner funksjon.
    
    x (float) er uavhengig variabel eller innvariabel.
    a og b (float) er parametre.
    y (float) er avhengig variabel eller utvariabel.
    """
    y = a*x + b
    return y

p0 = 2.5
p1 = 3.0
inn = 10.0

ut = fun_lin(inn, p0, p1)
print('Resultat = ', ut)
```

```python
help(fun_lin)
```

<h1 style="color:darkred;">Navnerom (namespace)</h1>

Nr vi lager funksjoner, er det nyttig  kjenne til begrepene **navnerom** (pythonsk: namespace) og **globale** og **lokale** variabler. Figuren nedenfor gir en illustrasjon. Begrepene er nrmere forklart under figuren.

* Variabler som er definert i hovedprogrammet, er globale variabler. De eksisterer i det globale navnerommet eller arbeidsomrdet (ogs kalt Python workspace).
* Variabler som er definert inne i en funksjon, er lokale variabler da de eksisterer kun i funksjonens navnerom. (De eksisterer ikke i det globale navnerommet.)
* Globale variabler kan leses, men i utgangspunktet ikke skrives (endres), inne i en funksjon.
* Globale variabler kan i utgangspunktet ikke oppdateres av kode som fins inne i en funksjon, men de kan allikevel oppdateres dersom nkkelordet global er brukt sammen med variabelen, f.eks. slik: global x global.

```python
def myfun():
    x_local = 20
    global x_global
    x_global += 1

x_global = 10

myfun()

print('x_global =', x_global)
# print('x_local =', x_local)
```

Hva skjer hvis vi prver  "printe ut" x_local (alts kjrer den kommenterte kodelinjen nederst i programmet ovenfor)? Fungerer? Feilmelding?

```python
def myfun():
    x_local = 20
    global x_global
    x_global += 1

x_global = 10

myfun()

print('x_global =', x_global)
print('x_local =', x_local)
```

<h1 style="color:darkred;">Moduler (modules)</h1>

<h2 style="color:darkgreen;">Hva er en modul?</h2>

En modul er et skript (*.py) som inneholder funksjoner og/eller variabler. Du kan bruke disse funksjonene og variablene i et hovedprogram etter at du har **importert** modulen inn til programmet (p samme mte som du tidligere i kurset har importert numpy og matplotlib). Figuren nedenfor illustrerer forholdet mellom en modul og hovedprogrammer.

<h2 style="color:darkgreen;">Hvorfor lage/bruke moduler?</h2>

Du kan samle (i moduler) funksjoner og variabler som du bruker i flere programmer  alts "flerbruksfunksjoner" og "flerbruksvariabler"  og dermed unng  ha kopier av dem i hvert av programmene. Dette gir mer effektiv programmering!

<h2 style="color:darkgreen;">Hvordan lager vi en modul?</h2>

Vi ser p et eksempel.

<ul>
<li>Modulens navn er min_modul.py
<li>Modulen (py-filen) er lagret i mappen modulmappe under mappen der hovedprogrammet ligger.
</ul>
